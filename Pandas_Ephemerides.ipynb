```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
dias = np.linspace(0,500,500)
data= pandas.read_csv("aaaa.csv",skiprows=2)
print(data["AU"])
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes()
#ax.set_facecolor("#e6e6e6")
plt.plot(dias, data["AU"].head(500), lw = 3)
plt.ylabel("Distancia ao Sol (AU)")
plt.xlabel("Dias")
plt.title("Efemerides de 4Vega")

plt.show()
```
